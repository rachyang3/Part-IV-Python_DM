# Market Basket Alalysis using Apriori

### Part 1: Data Preprocessing

```python
# import the Libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# import the dataset
dataset=pd.read_csv('dataset/Market_Basket_Optimisation.csv', header=None)
dataset.head()
```

```python
#read data in list
transactions=[]
for i in range(0,7501):
    transactions.append([str(dataset.values[i,j]) for j in range(0,20)])
```

```python
transactions[:5]
```

### Part 2: Training Apriori on the dataset

```python
# import readymate apriori like sklearn
from apyori import apriori
rules= apriori(transactions, min_support=0.003, min_confidence=0.2, min_lift=3, min_length=2)
```

### Part 3: Visualising  the Results

```python
# Check the rules
result=list(rules)
result[:30]
```
